```python
import numpy as np
import derpy
from time import sleep
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
from tqdm import tqdm
```

```python
cam = derpy.CRED2(set_temperature=-40, fps=10, tint=1, temp_tolerance=0.5) # TODO: add units to docstring (also add docstring)
```

```python
# Add Exposure times here
EXPOSURE_TIMES = [1,10,20,30,40,50,60,70,80,90,99] # Write exposure times (ms) here
EXPOSURE_TIMES = [1,10,99] # Write exposure times (ms) here
NUM_IMAGES = 10 # to take and median-combine
```

```python
lens_cap_off = []

for tint in tqdm(EXPOSURE_TIMES):
    cam.tint = tint
    img = cam.take_median_image(NUM_IMAGES)
    lens_cap_off.append(img)

```

Detector Noise With Lens Cap Off

As seen by the cell below, the detector noise is introduced somewhere between a 1-10ms integration time. This can be reproduced by rerunning the above and below cells with the lens cap removed. 
NOTE: The first graph below sometimes appears as straight 0 pixel values (purple image). Rerunning the camera a second time seems to fix the issue. 

```python
for i,j in zip(lens_cap_off, EXPOSURE_TIMES):
    plt.imshow(i, vmax=4000)
    plt.title(f"Integration Time (Lens Cap Removed): {j} ms")
    plt.colorbar()
    plt.show()
```

```python
lens_cap_on = []

for tint in tqdm(EXPOSURE_TIMES):
    cam.tint = tint
    img = cam.take_median_image(NUM_IMAGES)
    lens_cap_on.append(img)
```

```python
plt.figure()
for frame, tint in zip(lens_cap_off, EXPOSURE_TIMES):
    plt.hist(frame.ravel(), bins=256, range=(0.0, 4000), label=f"{tint} ms")
plt.legend()
plt.show()
```

Detector Noise With Lens Cap On


Here is the same process repeated with the lens cap on. 

```python
for i,j in zip(lens_cap_on, EXPOSURE_TIMES):
    plt.imshow(i, vmax=4000)
    plt.title(f"Integration Time (Lens Cap Attached): {j} ms")
    plt.colorbar()
    plt.show()
```
